```python
import pandas as pd
```

Plan on bringing in the last 6 months of 2017 (July - December) trip data csv files individually then combining them to have a csv with the mid-2017 trip data at the end

```python
# Read in July 2017 trip data csv
jul_df = pd.read_csv('201707-citibike-tripdata.csv')
# Show first 5 of July 2017 df
jul_df.head()
```

```python
# Read in August 2017 trip data csv
aug_df = pd.read_csv('201708-citibike-tripdata.csv')
# Show first 5 of August 2017 df
aug_df.head()
```

```python
# Read in September 2017 trip data csv
sep_df = pd.read_csv('201709-citibike-tripdata.csv')
# Show first 5 of September 2017 df
sep_df.head()
```

```python
# Read in October 2017 trip data csv
oct_df = pd.read_csv('201710-citibike-tripdata.csv')
# Show first 5 of October 2017 df
oct_df.head()
```

```python
# Read in November 2017 trip data csv
nov_df = pd.read_csv('201711-citibike-tripdata.csv')
# Show first 5 of November 2017
nov_df.head()
```

```python
# Read in December 2017 trip data csv
dec_df = pd.read_csv('201712-citibike-tripdata.csv')
# Show first 5 of December 2017
dec_df.head()
```

```python
# Create a new dataframe that uses concat to get 6 of the individual month data into mid 2017 data (July to December)
citibike_mid_2017_df = pd.concat([jul_df, aug_df, sep_df, oct_df, nov_df, dec_df])
# Show mid 2017 df
citibike_mid_2017_df
```

```python
# Create csv with July through December 2017 trip data from citibike_mid_2017_df
citibike_mid_2017_df.to_csv('citibike_mid_2017.csv', sep = ',')
```
